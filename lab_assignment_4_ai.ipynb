```python
"""
Lab Experiment 4: How to Use Git Effectively

Objectives:
1. Understand the purpose of version control systems.
2. Learn how to set up and configure Git.
3. Practice essential Git commands for managing code effectively.
4. Explore collaboration workflows with branching and merging.

Key Git Concepts:
- Repository (repo): A storage location for your project.
- Commit: A snapshot of your project at a certain time.
- Branch: A parallel version of your project for isolated development.
- Merge: Integrating changes from one branch into another.
- Remote: A version of your project hosted online (e.g., GitHub).
"""

def print_section(title):
    print("\n" + "=" * 60)
    print(title)
    print("=" * 60)

def step_1_install_configure():
    print_section("Step 1: Install and Configure Git")
    print("Check Git version:")
    print("  git --version\n")
    print("Configure username and email:")
    print('  git config --global user.name "Your Name"')
    print('  git config --global user.email "your.email@example.com"')

def step_2_create_repo():
    print_section("Step 2: Create a New Repository")
    print("Create project directory and initialize Git:")
    print("  mkdir my_project")
    print("  cd my_project")
    print("  git init")

def step_3_add_commit():
    print_section("Step 3: Add and Commit Files")
    print('Create a file:')
    print('  echo "Hello Git" > file.txt\n')
    print("Check repository status:")
    print("  git status\n")
    print("Stage the file:")
    print("  git add file.txt\n")
    print('Commit the file:')
    print('  git commit -m "Initial commit with file.txt"')

def step_4_branching():
    print_section("Step 4: Create and Switch Branches")
    print("Create a new branch:")
    print("  git branch new_feature\n")
    print("Switch to the new branch:")
    print("  git checkout new_feature")

def step_5_merge():
    print_section("Step 5: Merge Branches")
    print("Switch back to main branch:")
    print("  git checkout main\n")
    print("Merge new_feature into main:")
    print("  git merge new_feature")

def step_6_remote():
    print_section("Step 6: Connect to Remote Repository")
    print("Add remote repository:")
    print("  git remote add origin https://github.com/username/my_project.git\n")
    print("Push changes to remote:")
    print("  git push -u origin main")

def command_descriptions():
    print_section("Command Descriptions")
    commands = {
        "git init": "Initialize a repository",
        "git status": "Show changes status",
        "git add <file>": "Stage file changes",
        'git commit -m "msg"': "Save staged changes with a message",
        "git branch": "List branches",
        "git checkout <branch>": "Switch branches",
        "git merge <branch>": "Merge a branch into current branch",
        "git push": "Upload local changes to remote",
        "git pull": "Download changes from remote"
    }

    for cmd, desc in commands.items():
        print(f"{cmd:<25} : {desc}")

def main():
    print_section("Lab Experiment 4: How to Use Git Effectively")
    step_1_install_configure()
    step_2_create_repo()
    step_3_add_commit()
    step_4_branching()
    step_5_merge()
    step_6_remote()
    command_descriptions()

if __name__ == "__main__":
    main()
```
